```python
import pandas as pd
import sklearn as sk
import re
import string
import spacy
import nltk.corpus
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
stopwords_br = nltk.corpus.stopwords.words('portuguese')
stopwords_br
```

```python
df = pd.read_excel('FakeRecogna.xlsx')
df = df[['Titulo','Noticia','Classe']]
df
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df.isna().sum()
```

```python
df['Classe'] = df['Classe'].astype(int)
df
```

```python
df['News'] = df['Noticia']
df
```

```python
df = df[['News','Classe']]
df
```

```python
df[df['Classe']== 0].shape
```

```python
df[df['Classe']== 1].shape
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text) 
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)    
    return text
```

```python
df['News'] = df['News'].apply(wordopt)
df['News'] = df['News'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stopwords_br]))
df
```

```python
x = df['News']
y = df['Classe']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
xv_train = vectorizer.fit_transform(x_train)
xv_test = vectorizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
NB = MultinomialNB()
NB.fit(xv_train,y_train)
```

```python
pred_nb=NB.predict(xv_test)
```

```python
NB.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_nb))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"
    
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)

    nlp = spacy.load('pt_core_news_sm')
    doc = nlp(new_def_test["text"].iloc[0])
    lemmatized_text = " ".join([token.lemma_ for token in doc])
    new_x_test = [lemmatized_text]
    
    #new_x_test = new_def_test["text"]
    new_xv_test = vectorizer.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_NB = NB.predict(new_xv_test)
   

    return print("\n\nLR Prediction: {} \nNB Prediction: {}".format(output_lable(pred_LR[0]),  
                                                                    output_lable(pred_NB[0])))
```
