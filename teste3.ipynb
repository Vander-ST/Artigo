```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urlparse, urljoin
import time
import pandas as pd

# URL da primeira página
url = 'https://checamos.afp.com/list/all/37074/all/all/4525?page=0'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
link = soup.find('a', text='Última')
ultima_pagina = int(link['href'][-2:])+1 # converte para inteiro
i = 0

noticias = []

for i in range(i,ultima_pagina):

    url = f'https://checamos.afp.com/list/all/37074/all/all/4525?page={i}'

    # Faz a requisição HTTP
    response = requests.get(url)

    if response.status_code != 200:
        print(f'Site não está respondendo, status code: {response.status_code}')
        break

    # Analisa o HTML usando a biblioteca BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encontra a tag <main>
    main_tag = soup.find('main')

    # Encontra todas as tags <a> dentro da tag <main>
    a_tags = main_tag.find_all('a')

    # Itera sobre as tags <a> e extrai o conteúdo da tag <em>
    for a in a_tags:
        time.sleep(10) # De acordo com o arquio robots.txt do site, devemos esperar 10 segundos entre os requests
        href = a['href']
        if not href.startswith('http'):
            href = urljoin(url, href)
        link_response = requests.get(href)
        if response.status_code != 200:
            print(f'Site não está respondendo, status code: {link_response.status_code}')
            break
        link_soup = BeautifulSoup(link_response.text, 'html.parser')
        em_tag = link_soup.find('em')
        if em_tag:
            em_content = em_tag.text
            #print(em_content)
            noticias.append({'noticia': em_content, 'label': 'fake'})
            
    print(i)
    time.sleep(10)
df = pd.DataFrame(noticias)
df.to_excel('noticias.xlsx', index=False)
```
